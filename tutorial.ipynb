```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
pokemon = pd.read_csv('Pokemon.csv',index_col=0,encoding='latin')
pokemon.head()
```

```python
sns.lmplot(x='HP',y='Defense',data=pokemon)
```

```python
#🐈🐈🐈
sns.lmplot(x='Attack',y='HP', data = pokemon, fit_reg= False, hue = 'Stage')
```

```python
silly = pd.pivot_table(pokemon, index='Type 1', values='Attack', columns='Type 2', aggfunc='count').fillna(0)
silly
```

```python
sns.heatmap(silly, annot=True)
```
